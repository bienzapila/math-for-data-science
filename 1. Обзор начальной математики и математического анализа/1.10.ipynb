```python
from sympy import *

x, y = symbols('x y')
_y = x ** 2 + 1
z = y ** 3 - 2
```

```python
dz_dy = diff(z, y)
dy_dx = diff(_y, x)
dz_dx_chain = (dz_dy * dy_dx).subs(y, _y)
print(dz_dy, dy_dx, dz_dx_chain, sep='\n')
```

```python
dz_dx = diff(z.subs(y, _y))
print(dz_dx)
```

```python
print(dz_dx == dz_dx_chain)
```
